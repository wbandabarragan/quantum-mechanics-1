# Tutorial 10: Python for QM

## Special functions:

In class we studied the solutions to the Schrodinger equation in 3D and found a few interesting special functions:

- Legendre polynomials

- Legendre functions

- Laguerre polynomials

This tutorial will show you how to plot them using ``scipy``.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as spe
```

## 1. Legendre polynomials:

$P_{\ell}(x) \equiv \frac{1}{2^{\ell} \ell!} \left(\frac{d}{dx}\right)^{\ell} (x^2 - 1)^{\ell}$

```python
# Create a domain for x
x = np.linspace(-1., +1., 1000)
```

```python
# Empty list to store several p_l(x)
p_list = []

# Loop over l to produce several p_l(x)

for l in range(6):

  # Call scipy -> legendre
  p_l = spe.legendre(l)

  # Append the polynomial to the list
  p_list.append(p_l)
```

```python
# Double check
print(type(p_list[2]))
print("\n")
print(p_list[2])
print("\n")
display(p_list[2])
```

```python
# Plotting the Legendre polynomials

plt.figure(figsize=(9,4))

for l in range(6):
  plt.plot(x,p_list[l](x), label = rf"$\ell=${l}")

plt.xlabel(r"$x$")
plt.ylabel(r"$P_{\ell}(x)$")
plt.grid()
plt.legend()
plt.show()
```

## 2. Associated Legendre functions:

$P_{\ell}^m(x) \equiv (-1)^m \left( 1 - x^2 \right)^{m/2} \left( \frac{d}{dx} \right)^m P_{\ell}(x)$

$x=\cos(\theta)$

```python
# Create a vector for theta
theta = np.linspace(0., 2*np.pi, 1000)

# Create a vector for x
x = np.cos(theta)

# Fix the input quantum numbers
l = 2

# Empty list to store the associated legendre functions
p_lm = []

# Loop over m (we restrict ourselves to + m values)
for m in range(0, l+1):
  #print(m)

  p_lm.append(spe.lpmv(m, l, x))

print(len(p_lm))
```

### Norm of the associated Legendre function
 $$r=|P_{\ell}^m(\cos(\theta))|$$

```python
# Plotting in the polar angle coordinate -> theta
fig, ax = plt.subplots(1, 1, figsize=(4,4), subplot_kw = {"projection": "polar"})

for m in range(0, l+1):
  # Compute r
  r = np.abs(p_lm[m])
  plt.plot(theta, r, label = rf"$(\ell, m)=$({l},{m})")

ax.set_theta_offset(np.pi/2.)
plt.legend()
plt.show()
```

## 3. Laguerre polynomials:

$L_q(x) \equiv \frac{e^{x}}{q!}\left(\frac{d}{dx}\right)^q\left(e^{-x}x^q\right)$


```python
# Create a new x axis
x = np.linspace(0., 10., 1000)

# Empty list to store several L_q(x)
lag_list = []

# Loop over l to produce several L_q(x)

for q in range(6):

  # Call scipy -> legendre
  l_q = spe.laguerre(q)

  # Append the polynomial to the list
  lag_list.append(l_q)
```

```python
# Double check
#print(lag_list)
```

```python
# Plotting the Legendre polynomials

plt.figure(figsize=(9,4))

for q in range(6):
  plt.plot(x,lag_list[q](x), label = rf"$q=${q}")

plt.xlabel(r"$x$")
plt.ylabel(r"$L_{q}(x)$")
plt.grid()

plt.ylim(-10, +10)
plt.legend()
plt.show()
```

### On the role of the decreasing exponential term:

```python
# Plotting the Legendre polynomials

plt.figure(figsize=(9,4))

for q in range(6):
  plt.plot(x,lag_list[q](x)*np.exp(-0.7*x), label = rf"$q=${q}")

plt.xlabel(r"$x$")
plt.ylabel(r"$L_{q}(x)$")
plt.grid()

plt.ylim(-3, +3)
plt.legend()
plt.show()
```
