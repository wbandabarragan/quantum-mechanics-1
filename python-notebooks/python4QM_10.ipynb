## Special functions:

In class we studied the solutions to the Schrodinger equation in 3D and found a few interesting special functions:

- Legendre polynomials

- Legendre functions

- Laguerre polynomials

This tutorial will show you how to plot them using ``scipy``.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as spe
```

## 1. Legendre polynomials:

$P_{\ell}(x) \equiv \frac{1}{2^{\ell} \ell!} \left(\frac{d}{dx}\right)^{\ell} (x^2 - 1)^{\ell}$

```python
# Create a domain for x
x = np.linspace(-1., +1., 1000)
```

```python
# Empty list to store several p_l(x)
p_list = []

# Loop over l to produce several p_l(x)

for l in range(6):

  # Call scipy -> legendre
  p_l = spe.legendre(l)

  # Append the polynomial to the list
  p_list.append(p_l)
```

```python
# Double check
print(type(p_list[3]))
print("\n")
print(p_list[0])
print("\n")
display(p_list[3])
```

```python
# Plotting the Legendre polynomials

plt.figure(figsize=(9,4))

for l in range(6):
  plt.plot(x,p_list[l](x), label = rf"$\ell=${l}")

plt.xlabel(r"$x$")
plt.ylabel(r"$P_{\ell}(x)$")
plt.grid()
plt.legend()
plt.show()
```
