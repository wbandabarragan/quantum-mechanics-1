## 3. Vector generation:

- ``np.arange()``
- ``np.linspace()``


### Example:
We can use Latex on a Colab notebook (in markdown cells).

$$\Psi(x,t) = e^{ikx-i\omega t}$$

The spatial part:

$$\psi(x) = e^{ikx}=\cos(kx)+i\sin(kx)$$

Taking the Real part:

$${\rm I\!R}(\psi(x))=cos(kx)$$

where: $k$ is the wavenumber (with units of 1/length).


```python
# Vectors for the X and Y axes

# Use spacing between points in the vector
x = np.arange(-3*np.pi, +3*np.pi, 0.01)

# Define the wavenumber/spatial frequency
k1 = 1.0
k2 = 2.5
k3 = 5.0

# Evalue the Y axis
y1 = np.cos(k1*x)
y2 = np.cos(k2*x)
y3 = np.cos(k3*x)
```

## 4. 1D Plotting:

```python
# Plotting Environment
plt.figure(figsize=(9,4))

# Plot customisation
plt.plot(x, y1, color = "red", linestyle = "-.", label = r"$y_1 = \cos(k_1x)$")
plt.plot(x, y2, color = "blue", linestyle = "-", label = r"$y_2 = \cos(k_2x)$")
#plt.plot(x, y3, color = "green", linestyle = ":", label = r"$y_3 = \cos(k_3x)$")

# Add the legend
plt.legend(loc=1)

# Customise the axes labels
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")

# Add a title
plt.title("Cosine Functions")

# Add axes limits
plt.xlim(-3*np.pi, +3*np.pi)
plt.ylim(-1.1, +1.5)

# Add a grid
plt.grid()

plt.show()
```

## 5. 2D Functions:

Now we want to represent a 2D particle:

$$\psi(x,y)=\cos(k_x\,x) + \sin(k_y\,y)$$

where: $k_x$ and $k_y$ are wavenumbers along X and Y (with units of 1/length).



```python
# To represent 2D surfaces, we need a meshgrid

# Create 1D vectors for the axes
x = np.linspace(-3*np.pi, +3*np.pi, 1000)
y = np.linspace(-3*np.pi, +3*np.pi, 1000)

# Now we create the 2D meshgrid
x_2d, y_2d = np.meshgrid(x, y)

#print(x.shape, x_2d.shape)
#print(y.shape, y_2d.shape)
```

$$\psi(x,y)=\cos(k_x\,x) + \sin(k_y\,y)$$

```python
# Python Function
def psi_2d(x_2d, y_2d):
  """
  This is what we call a docstring.
  Inputs: x_2d, y_2d
  Output: psi_2d
  """
  psi_2d = np.cos(k_x*x_2d) + np.sin(k_y*y_2d)

  return psi_2d
```

```python
# Define the wavenumber
k_x = 2.
k_y = 1.

# We call the funtion
psi_xy = psi_2d(x_2d, y_2d)

#print(psi_xy.shape)
```

### 6. 2D surfaces in 3D

```python
# Plotting: Figure Environment
fig = plt.figure(figsize=(8,8))

# Axis object
ax = fig.add_subplot(111, projection = "3d")

# Plot
surf = ax.plot_surface(x_2d, y_2d, psi_xy, cmap = "inferno")

# Add the colour bar
#fig.colorbar(surf)

# Add labels
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
ax.set_zlabel(r"$z$")

# This may help
#plt.tight_layout()

# Show plot
plt.show()
```

### 7. 2D surfaces on the XY plane


``plt.pcolor``: this is an interpolation function

```python
plt.figure(figsize=(6,4))

surf = plt.pcolor(x_2d, y_2d, psi_xy, cmap = "inferno", clim = (-2, 2))

plt.colorbar(surf)

plt.show()
```
