## 3. Vector generation:

- ``np.arange()``
- ``np.linspace()``


### Example:
We can use Latex on a Colab notebook (in markdown cells).

$$\Psi(x,t) = e^{ikx-i\omega t}$$

The spatial part:

$$\psi(x) = e^{ikx}=\cos(kx)+i\sin(kx)$$

Taking the Real part:

$${\rm I\!R}(\psi(x))=cos(kx)$$

where: $k$ is the wavenumber (with units of 1/length).


```python
# Vectors for the X and Y axes

# Use spacing between points in the vector
x = np.arange(-3*np.pi, +3*np.pi, 0.01)

# Define the wavenumber/spatial frequency
k1 = 1.0
k2 = 2.5
k3 = 5.0

# Evalue the Y axis
y1 = np.cos(k1*x)
y2 = np.cos(k2*x)
y3 = np.cos(k3*x)
```

## 4. Plotting:

```python
# Plotting Environment
plt.figure(figsize=(9,4))

# Plot customisation
plt.plot(x, y1, color = "red", linestyle = "-.", label = r"$y_1 = \cos(k_1x)$")
plt.plot(x, y2, color = "blue", linestyle = "-", label = r"$y_2 = \cos(k_2x)$")
#plt.plot(x, y3, color = "green", linestyle = ":", label = r"$y_3 = \cos(k_3x)$")

# Add the legend
plt.legend(loc=1)

# Customise the axes labels
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")

# Add a title
plt.title("Cosine Functions")

# Add axes limits
plt.xlim(-3*np.pi, +3*np.pi)
plt.ylim(-1.1, +1.5)

# Add a grid
plt.grid()

plt.show()
```
